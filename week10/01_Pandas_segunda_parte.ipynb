# Subset de un dataframe
Es obtener un dataframe m√°s peque√±o de n filas y/o m columnas para poderlas manipular de manera m√°s eficiente.

````python
df.mag # obtenemos los valores de un solo atributo
````

````python
df['mag'] # es lo mismo que el m√©todo anterior
````

````python
# Seleccionaremos solo 3 columnas
df[['mag', 'alert', 'title']]
````

````python
# Equivalente al m√©todo anterior ser√≠a declarar las columnas previamente
cols = ['mag', 'alert', 'title']
df[cols]
````

````python
# Tambi√©n podemos utilizar las list comprehension
df[
   ['title', 'alert', 'rms'] +
   [col for col in df.columns if col.startswith('mag')]
]
````

````python
# Poder analizar solo unas determinadas columnas
cols_c = [col for col in df.columns if col.startswith('c')]
cols_c
````

````python
# eXplicando el c√≥digo de arriba y a√±anidos un separador de l√≠nea \
['title','time'] \
+ [col for col in df.columns if col.startswith('mag')] # el resultado obtenido
````

````python
# Realizamos operaciones de slicing
df[10:100] # mostrar√° las filas desde el √≠ndice 10 hasta el 100-1 (99)
````

````python
# Filtramos por columnas y filas
df[['title', 'mag']][50:101]
````

````python
# La alternativa
cols = ['title', 'mag']
df[cols][50:101] # es equivalente a df[['title', 'mag']][50:101]
````

````python
# El orden no importa
df[50:101][cols].equals(df[cols][50:101])
````

# Realizar subset de datos con operaciones de indexing: `iloc` y `loc`

````python
# La funci√≥n loc[] utilizar√° llamando los nombres de filas y columnas
subset_1 = df.loc[10:101, 'title'] # el slice start y end completos.
subset_1
````

````python
# Observamos el tipo de objeto creado
type(subset_1)
````

````python
df.loc[110:112, 'title'] = df.loc[110:112, 'title'].str.lower()
df.loc[110:112, 'title']
````

````python
df.loc[:,'mag']
````

````python
# Si deseamos a√±adir m√°s columnas
subset_2 = df.loc[10:15, ['title','mag','time']]
subset_2
````

````python
cols = ['title','mag','time']
subset_3 = df.loc[10:15, cols]
subset_3
````

````python
subset_2.equals(subset_3)
````

````python
# M√©todo con iloc, que ser√≠a b√∫squeda por posici√≥n
df.iloc[10:15, [19, 8]] # columnas pos 19 y pos 8
````

````python
# Realizar una slice de columnas
df.iloc[50:101, 6:11] # omite la √∫ltima fila y la √∫ltima columna
````

````python
# Podemos evaluar los dos m√©todos loc e iloc
df.iloc[10:15, 6:10].equals(
    df.loc[10:14, 'gap':'magType']
)
````

# Uso de filtros

````python
# Si a√±adimos una condici√≥n
df.mag > 3
````

````python
# El m√©todo anterior en el contexto dataframe entero
df[df.mag > 3] # devolver√° n filas cuando la condici√≥n sea True o 1
````

````python
# Utilizamos loc y la condici√≥n para crear un nuevo subset
subset_4 = df.loc[
                  df.mag >= 7.,
                  ['alert','mag','magType','title','type']
]
subset_4
````
