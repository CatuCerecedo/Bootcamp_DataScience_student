```python
# Importar librerías
import numpy as np
import pandas as pd
import datetime as dt
```

```python
# Cargamos archivo de github
ruta = 'https://raw.githubusercontent.com/CatuCerecedo/Bootcamp_DataScience_student/master/week10/data/parsed.csv'
df = pd.read_csv(ruta)
df.info()
```

```python
df.head()
```

```python
# Percentil 95 de la magnitud del terremoto en Japón usando el tipo de magnitud `mb`
# p95 = df[(df['magType'] == 'mb') & (df['parsed_place'] == 'Japan') & (df['type'] == 'earthquake')]['mag'].describe(percentiles=[0.05, 0.95])[6]
p95 = df[(df['magType'] == 'mb') & (df['parsed_place'] == 'Japan')]['mag'].describe(percentiles=[0.05, 0.95])[6]
print(f'Percentil 95 de la magnitud del terremoto en Japón usando el tipo de magnitud `mb`: {p95}')

#... Alternativa con método quantile de pd
# df[(df['magType'] == 'mb') & (df['parsed_place'] == 'Japan')]['mag'].quantile(0.95)
```

```python
# Encuentre el porcentaje de terremotos en Indonesia que se acompañaron de tsunamis.
# ¿Teniendo en cuenta Indonesia como el total?
total_indonesia = len(df[(df['type'] == 'earthquake') & (df['parsed_place'] == 'Indonesia')])
tsunami = len(df[(df['type'] == 'earthquake') & (df['parsed_place'] == 'Indonesia') & (df['tsunami'] == 1)])
print(f'El total de terremotos en indonesia fue de {total_indonesia}, de los cuales con tsunami se contabilizaron {tsunami}.\n El procentaje de terremotos con tsunami fue de {round(tsunami/total_indonesia*100, 2)}%')
```

```python
#...Alternativa
f"{df[df['parsed_place'] == 'Indonesia']['tsunami'].value_counts(normalize=True).iloc[1,]:.2%}"
```

```python
# Calcule estadísticas resumidas para terremotos en Nevada.
df[(df['type'] == 'earthquake') & (df['parsed_place'] == 'Nevada')].describe()
```

```python
# Agregue una columna que indique si el terremoto ocurrió en un país o estado de EE. UU. Que se encuentra en el "Anillo de fuego". 
# Utilice Alaska, Antarctica (buscar por Antarctic), Bolivia, California, Canada, Chile, Costa Rica, Ecuador, Fiji, 
# Guatemala, Indonesia, Japan, Kermadec Islands, Mexico (no es New Mexico), New Zealand, Peru, Philippines, Russia, Taiwan, Tonga, and Washington.
cols = ['Alaska', 'Antarctic', 'Bolivia', 'California', 'Canada', 'Chile', 'Costa Rica', 'Ecuador', 'Fiji', 
 'Guatemala', 'Indonesia', 'Japan', 'Kermadec Islands', 'Mexico', 'New Zealand', 'Peru', 'Philippines', 'Russia', 'Taiwan', 'Tonga', 'Washington']

df['ring fire'] = df['parsed_place'].isin(cols) 
df.sample(20, random_state=17)
```

```python
#...Alternativa. Cuando quieres buscar patrones
df['alter'] = df['parsed_place'].str.contains(r'|'.join([
    'Alaska', 'Antarctic', 'Bolivia', 'California', 'Canada', 'Chile', 'Costa Rica', 'Ecuador', 'Fiji', 
 'Guatemala', 'Indonesia', 'Japan', 'Kermadec Islands', '^Mexico', 'New Zealand', 'Peru', 'Philippines', 
 'Russia', 'Taiwan', 'Tonga', 'Washington'
]))
```

```python
# Calcula el número de terremotos en las ubicaciones del Anillo de Fuego y el número fuera de ellos.
eq_out = len(df[(df['type'] == 'earthquake') & (df['ring fire'] == False)])
eq_in =  len(df[(df['type'] == 'earthquake') & (df['ring fire'] == True)])

print(f'Número de terremotos:\n - Dentro del anillo: {eq_in}\n - Fuera del anillo: {eq_out}')
```

```python
#...Alternativa
df[(df['type'] == 'earthquake')]['ring fire'].value_counts()
```

```python
# Encuentra el recuento de tsunamis a lo largo del Anillo de fuego.
tsunami_ring = len(df[(df['ring fire'] == True) & (df['tsunami'] == 1)])
print(f'Se produjeron un total de {tsunami_ring} tsunamis en el Anillo de Fuego')
```
